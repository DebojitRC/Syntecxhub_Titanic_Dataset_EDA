```python
## Objective
# To analyze survival patterns in the Titanic dataset using data analysis and visualization techniques.
```

```python
## Import Libraries
# Pandas is used for data handling, while Seaborn and Matplotlib are used for visualization.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Load Dataset
# The dataset is loaded into a Pandas DataFrame for analysis.

df = pd.read_csv("train.csv")
df.head()
```

```python
## Inspect Dataset
# This step helps understand the structure of the dataset.

df.columns
```

```python
df.info()
```

```python
## Check Missing Values
# Identifies missing values in the dataset.

df.isnull().sum()
```

```python
## Survival Distribution
# Shows how many passengers survived vs not survived.

df['Survived'].value_counts()
```

```python
sns.countplot(x='Survived', data=df)

plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")

plt.savefig("survival_count.png")
plt.show()
```

```python
## Survival by Gender
# Compares survival rates between male and female passengers.

sns.countplot(x='Sex', hue='Survived', data=df)

plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.savefig("survival_gender.png")
plt.show()
```

```python
## Survival by Passenger Class
# Analyzes how passenger class affected survival.

sns.countplot(x='Pclass', hue='Survived', data=df)

plt.title("Survival by Passenger Class")
plt.xlabel("Class")
plt.ylabel("Count")

plt.savefig("survival_class.png")
plt.show()
```

```python
## Age Group Analysis
# Passengers are divided into age groups to analyze survival patterns.

df['AgeGroup'] = pd.cut(df['Age'],
                       bins=[0,12,19,59,100],
                       labels=['Child','Teen','Adult','Senior'])

df[['Age','AgeGroup']].head()
```

```python
sns.countplot(x='AgeGroup', hue='Survived', data=df)

plt.title("Survival by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")

plt.savefig("survival_agegroup.png")
plt.show()
```

```python
## Age Distribution (Boxplot)
# Shows age distribution among survivors and non-survivors.

sns.boxplot(x='Survived', y='Age', data=df)

plt.title("Age Distribution by Survival")
plt.xlabel("Survived")
plt.ylabel("Age")

plt.savefig("age_distribution.png")
plt.show()
```

```python
## Insights

# Female passengers show a much higher survival rate compared to males
# First-class passengers have better survival chances than lower classes
# Third-class passengers have the highest number of deaths
# Children tend to have relatively higher survival rates
# The overall number of non-survivors is greater than survivors
```
